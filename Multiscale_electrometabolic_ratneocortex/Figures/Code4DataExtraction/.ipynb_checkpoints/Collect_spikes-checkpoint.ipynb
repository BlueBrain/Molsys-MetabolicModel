```python
def read_result(result_path):
    read_result = h5py.File(result_path)
    result = read_result['report']['All']['data']

    return(result)

def mapping_extract(result_path):
    read_result = h5py.File(result_path)
    result = read_result['report']['All']['data']
        
    extract_mapping = read_result['report']['All']['mapping']['node_ids'][:]
    
    t0 = read_result['report']['All']['mapping']['time'][0]
    T =read_result['report']['All']['mapping']['time'][1]
    dt = read_result['report']['All']['mapping']['time'][2]

    
    N= (T-t0)/dt
    time_line = np.linspace(t0,T, int(N))
    N2 = (T-t0)/0.025
    voltage_time_line = np.linspace(t0,T,int(N2))
    return(extract_mapping, time_line, voltage_time_line)
```

```python
sim_results_path = results_path + 'reporting_3000_node_sets_v03_single_node_removed/'

ATP_met = read_result(sim_results_path+  'ndam_atpi.h5')
after_ATP_met = read_result(sim_results_path+ 'metab_atpi_after_sync.h5')
```

```python
# Assuming df is your DataFrame and col_prod, col_cons are defined colors for production and consumption
layers = np.unique(df['layer'])

# Create a figure and axes for the plot
fig, axes = plt.subplots(figsize=(8, 6))

# Set the bar width
bar_width = 0.35
x = np.arange(len(layers))  # the label locations

# Create empty lists to store counts
count_prod = []
count_cons = []

# Count the number of entries for each layer
for layer in layers:
    count_prod.append((df[df.layer == layer]["spikes_met"]).sum())
    count_cons.append((df[df.layer == layer]["spikes_ndam"]).sum())
    print(count_prod, count_cons)

# Create bar plots
bars_prod = axes.bar(x + bar_width / 2, count_prod, width=bar_width, color='tab:orange', label='Spike Young')
bars_cons = axes.bar(x - bar_width / 2, count_cons, width=bar_width, color='tab:blue', label='Spike Ndam')

# Set x-ticks and labels
axes.set_xticks(x)
axes.set_xticklabels(layers, rotation=30)
axes.set_xlabel('Layer', fontsize=12)
axes.set_ylabel('Count', fontsize=12)

# Add legend
axes.legend(loc='upper left')

# Hide right and top spines for a cleaner look
axes.spines[['right', 'top']].set_visible(False)

# Tight layout and show the plot
fig.tight_layout()
plt.show()

# Optional: Save the plot
resolution_value = 1200
# plt.savefig("./RESULTS/Etype_ATP_cons_pro_mM_s_count_barplot.pdf", format="pdf", dpi=resolution_value, bbox_inches='tight')
```
